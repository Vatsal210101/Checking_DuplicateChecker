```python
# Advanced Features
from nltk.corpus import stopwords

def fetch_token_features(row):

    q1 = row['question1']
    q2 = row['question2']

    SAFE_DIV = 0.0001

    STOP_WORDS = stopwords.words("english")

    token_features = [0.0]*8

    # Converting the Sentence into Tokens:
    q1_tokens = q1.split()
    q2_tokens = q2.split()

    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return token_features

    # Get the non-stopwords in Questions
    q1_words = set([word for word in q1_tokens if word not in STOP_WORDS])
    q2_words = set([word for word in q2_tokens if word not in STOP_WORDS])

    #Get the stopwords in Questions
    q1_stops = set([word for word in q1_tokens if word in STOP_WORDS])
    q2_stops = set([word for word in q2_tokens if word in STOP_WORDS])

    # Get the common non-stopwords from Question pair
    common_word_count = len(q1_words.intersection(q2_words))

    # Get the common stopwords from Question pair
    common_stop_count = len(q1_stops.intersection(q2_stops))

    # Get the common Tokens from Question pair
    common_token_count = len(set(q1_tokens).intersection(set(q2_tokens)))


    token_features[0] = common_word_count / (min(len(q1_words), len(q2_words)) + SAFE_DIV)
    token_features[1] = common_word_count / (max(len(q1_words), len(q2_words)) + SAFE_DIV)
    token_features[2] = common_stop_count / (min(len(q1_stops), len(q2_stops)) + SAFE_DIV)
    token_features[3] = common_stop_count / (max(len(q1_stops), len(q2_stops)) + SAFE_DIV)
    token_features[4] = common_token_count / (min(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    token_features[5] = common_token_count / (max(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)

    # Last word of both question is same or not
    token_features[6] = int(q1_tokens[-1] == q2_tokens[-1])

    # First word of both question is same or not
    token_features[7] = int(q1_tokens[0] == q2_tokens[0])

    return token_features
```

```python
import distance

def fetch_length_features(row):

    q1 = row['question1']
    q2 = row['question2']

    length_features = [0.0]*3

    # Converting the Sentence into Tokens:
    q1_tokens = q1.split()
    q2_tokens = q2.split()

    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return length_features

    # Absolute length features
    length_features[0] = abs(len(q1_tokens) - len(q2_tokens))

    #Average Token Length of both Questions
    length_features[1] = (len(q1_tokens) + len(q2_tokens))/2

    strs = list(distance.lcsubstrings(q1, q2))
    length_features[2] = len(strs[0]) / (min(len(q1), len(q2)) + 1)

    return length_features
```

```python
import numpy as np
from scipy.sparse import hstack, csr_matrix

def _safe_div(a, b):
    return float(a) / float(b) if b else 0.0

def query_point_creator(q1, q2, *, vectorizer=None):
    """
    Builds a single-row sparse feature vector:
    [handcrafted features | TF-IDF(q1) | TF-IDF(q2)]
    Assumes the helper functions below return iterables of numbers:
      - test_fetch_token_features(q1, q2)
      - test_fetch_length_features(q1, q2)
      - test_fetch_fuzzy_features(q1, q2)
      - test_common_words(q1, q2)
      - test_total_words(q1, q2)
    And that `preprocess` returns a string.
    """
    # use provided vectorizer or fall back to a global `tfidf`
    vec = vectorizer if vectorizer is not None else tfidf

    # preprocess
    q1 = preprocess(str(q1))
    q2 = preprocess(str(q2))

    # basic counts
    len_q1_chars = len(q1)
    len_q2_chars = len(q2)
    len_q1_tokens = len(q1.split())
    len_q2_tokens = len(q2.split())

    # word overlap features
    cw = test_common_words(q1, q2)
    tw = test_total_words(q1, q2)
    cw_ratio = round(_safe_div(cw, tw), 2)

    # collect handcrafted features
    features = [
        len_q1_chars, len_q2_chars,
        len_q1_tokens, len_q2_tokens,
        cw, tw, cw_ratio
    ]

    # extend with your other feature blocks (ensure they return flat lists/1D arrays)
    features += list(test_fetch_token_features(q1, q2))
    features += list(test_fetch_length_features(q1, q2))
    features += list(test_fetch_fuzzy_features(q1, q2))

    # convert handcrafted features to a 1xN sparse row
    dense_row = np.asarray(features, dtype=float).reshape(1, -1)
    handcrafted_sparse = csr_matrix(dense_row)

    # vector features (TF-IDF)
    q1_vec = vec.transform([q1])   # 1 x V
    q2_vec = vec.transform([q2])   # 1 x V

    # final sparse row: [handcrafted | q1_vec | q2_vec]
    X_row = hstack([handcrafted_sparse, q1_vec, q2_vec])
    return X_row
```

```python
# ==== SAFETY/UTILITY PATCH BLOCK ====
import re
import numpy as np
from scipy.sparse import hstack, csr_matrix

# --- stopwords (safe import) ---
try:
    from nltk.corpus import stopwords
    STOP_WORDS = set(stopwords.words("english"))
except Exception:
    STOP_WORDS = {
        "the","a","an","and","or","in","on","for","to","of","is","are","was","were","be","been","by","with","as","at","from"
    }

# --- fuzzy (RapidFuzz preferred; fallback to fuzzywuzzy; final mini fallback) ---
try:
    from rapidfuzz import fuzz
except Exception:
    try:
        from fuzzywuzzy import fuzz
    except Exception:
        def _norm_ws(s): return " ".join(str(s).lower().split())
        class _MiniFuzz:
            @staticmethod
            def QRatio(a,b):
                a=_norm_ws(a); b=_norm_ws(b)
                return int(100 * (len(set(a.split()) & set(b.split()))) / (len(set(a.split()) | set(b.split())) or 1))
            partial_ratio = token_sort_ratio = token_set_ratio = QRatio
        fuzz = _MiniFuzz()

# --- preprocess (simple) ---
def preprocess(s: str) -> str:
    s = str(s).lower()
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

# --- helpers you *must* have defined ---
def test_common_words(q1,q2):
    w1 = set(map(lambda w: w.lower().strip(), q1.split()))
    w2 = set(map(lambda w: w.lower().strip(), q2.split()))
    return len(w1 & w2)

def test_total_words(q1, q2):
    w1 = set(map(lambda w: w.lower().strip(), q1.split()))
    w2 = set(map(lambda w: w.lower().strip(), q2.split()))
    return len(w1) + len(w2)

def _safe_div(a, b):
    return float(a) / float(b) if b else 0.0

# --- longest common substring length (pure Python; no 'distance' package) ---
def _lcs_len(a: str, b: str) -> int:
    m, n = len(a), len(b)
    if m == 0 or n == 0:
        return 0
    dp = [0]*(n+1)
    best = 0
    for i in range(1, m+1):
        prev = 0
        ai = a[i-1]
        for j in range(1, n+1):
            tmp = dp[j]
            if ai == b[j-1]:
                dp[j] = prev + 1
                if dp[j] > best: best = dp[j]
            else:
                dp[j] = 0
            prev = tmp
    return best

def test_fetch_token_features(q1,q2):
    SAFE_DIV = 1e-4
    q1_tokens = q1.split(); q2_tokens = q2.split()
    feats = [0.0]*8
    if not q1_tokens or not q2_tokens:
        return feats
    q1_words = set([w for w in q1_tokens if w not in STOP_WORDS])
    q2_words = set([w for w in q2_tokens if w not in STOP_WORDS])
    q1_stops = set([w for w in q1_tokens if w in STOP_WORDS])
    q2_stops = set([w for w in q2_tokens if w in STOP_WORDS])
    common_word_count  = len(q1_words & q2_words)
    common_stop_count  = len(q1_stops & q2_stops)
    common_token_count = len(set(q1_tokens) & set(q2_tokens))
    feats[0] = common_word_count / (min(len(q1_words),  len(q2_words))  + SAFE_DIV)
    feats[1] = common_word_count / (max(len(q1_words),  len(q2_words))  + SAFE_DIV)
    feats[2] = common_stop_count / (min(len(q1_stops),  len(q2_stops))  + SAFE_DIV)
    feats[3] = common_stop_count / (max(len(q1_stops),  len(q2_stops))  + SAFE_DIV)
    feats[4] = common_token_count / (min(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    feats[5] = common_token_count / (max(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    feats[6] = int(q1_tokens[-1] == q2_tokens[-1])
    feats[7] = int(q1_tokens[0]  == q2_tokens[0])
    return feats

def test_fetch_length_features(q1,q2):
    feats = [0.0]*3
    q1_tokens = q1.split(); q2_tokens = q2.split()
    if not q1_tokens or not q2_tokens:
        return feats
    feats[0] = abs(len(q1_tokens) - len(q2_tokens))
    feats[1] = (len(q1_tokens) + len_q2_tokens) / 2.0 if (len_q2_tokens := len(q2_tokens)) or True else 0.0  # safe average
    lcs = _lcs_len(q1, q2)
    feats[2] = lcs / (min(len(q1), len(q2)) + 1.0)
    return feats

def test_fetch_fuzzy_features(q1,q2):
    return [
        fuzz.QRatio(q1, q2),
        fuzz.partial_ratio(q1, q2),
        fuzz.token_sort_ratio(q1, q2),
        fuzz.token_set_ratio(q1, q2),
    ]

# --- final builder (expects a fitted TF-IDF vectorizer named `tfidf`, or pass vectorizer=...) ---
def query_point_creator(q1, q2, *, vectorizer=None):
    vec = vectorizer if vectorizer is not None else tfidf  # ensure 'tfidf' exists
    q1 = preprocess(str(q1)); q2 = preprocess(str(q2))
    len_q1_chars = len(q1); len_q2_chars = len(q2)
    len_q1_tokens = len(q1.split()); len_q2_tokens = len(q2.split())
    cw = test_common_words(q1, q2)
    tw = test_total_words(q1, q2)
    cw_ratio = round(_safe_div(cw, tw), 2)
    features = [
        len_q1_chars, len_q2_chars,
        len_q1_tokens, len_q2_tokens,
        cw, tw, cw_ratio
    ]
    features += list(test_fetch_token_features(q1, q2))
    features += list(test_fetch_length_features(q1, q2))
    features += list(test_fetch_fuzzy_features(q1, q2))
    handcrafted_sparse = csr_matrix(np.asarray(features, dtype=float).reshape(1, -1))
    q1_vec = vec.transform([q1])
    q2_vec = vec.transform([q2])
    return hstack([handcrafted_sparse, q1_vec, q2_vec])
# ==== END PATCH BLOCK ====
```
